```python
from scipy.stats import poisson
import numpy as np
import matplotlib.pyplot as plt
```

```python
prob_even = []
for k in range(102):
    if k % 2 == 0:
        prob_even.append(poisson.pmf(k,20))
    else:
        prob_even.append(0)
```

```python
plt.figure(figsize=(8,6))
plt.plot(range(102), np.cumsum(prob_even))
plt.xlabel('k', fontsize=12)
plt.ylabel('Cumulative sum of prob. of selling k (even) cakes', fontsize=12)
```
