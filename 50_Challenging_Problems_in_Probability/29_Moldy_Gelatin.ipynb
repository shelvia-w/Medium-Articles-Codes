### Moldy Gelatin

```python
import matplotlib.pyplot as plt
import math
```

```python
m = range(1,100)
p = []
for val in m:
    p.append(math.exp(-val)*(val)**val/math.factorial(val))
plt.plot(m,p, label=r'$\frac{e^{-m}(m)^{m}}{m!}$')
plt.xlabel('m', fontsize=12)
plt.ylabel('prob. of having exactly m colonies', fontsize=12)
plt.legend(fontsize=20)
plt.show()
```
