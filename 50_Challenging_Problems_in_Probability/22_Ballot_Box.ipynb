### Ballot Box

```python
from itertools import permutations
import matplotlib.pyplot as plt
```

```python
n_a = 3
n_b = 2
tally_list = ['A']*n_a+['B']*n_b
all_sequences = set(list(permutations(tally_list)))

tie = 0
total = 0
for s in all_sequences:
    a = 0
    b = 0
    for i in s:
        if i == 'A':
            a += 1
        else:
            b += 1
        if a == b:
            tie += 1
            break
    total += 1
print(f'Probability of having same no. of tallies is {tie}/{total}.')
        
```

```python
all_p_list = []
for n_a in range(2,6):
    p_list = []
    for n_b in range(1,n_a+1):
        tally_list = ['A']*n_a+['B']*n_b
        all_sequences = set(list(permutations(tally_list)))
        tie = 0
        total = 0
        for s in all_sequences:
            a = 0
            b = 0
            for i in s:
                if i == 'A':
                    a += 1
                else:
                    b += 1
                if a == b:
                    tie += 1
                    break
            total += 1
        p_list.append(tie/total)
    all_p_list.append(p_list)
```

```python
for i, n_a in enumerate(range(2,6)):
    plt.plot(range(1,n_a+1), all_p_list[i], label=f'a = {n_a}', marker='o')
plt.legend(fontsize=12)
plt.grid(True)
plt.xlabel('b', fontsize=12)
plt.ylabel('prob. of having same no. of tallies', fontsize=12)
plt.xticks(range(1,n_a+1))
plt.show()
```
